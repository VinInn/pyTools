```python
from mpmath import mp, taylor, chebyfit, cos, nprint, polyval, log, log1p
import numpy as np
```

```python
mp.pretty = True
```

```python
poly, err = chebyfit(log, [0.75, 1.5], 10, error=True)
nprint(poly,12)
nprint(err,12)
error = lambda x: abs(log(x) - polyval(poly, x))
nprint(max([error(0.75*(1.+n/1000.)) for n in range(1000)]), 12)
```

```python
def f(x) :
    return log1p(x)/x
nprint(taylor(log1p, 0, 10))
poly, err = chebyfit(f, [-0.25, 0.5], 9, error=True)
nprint(poly,12)
nprint(err,12)
```

```python
error = lambda x: abs(log(x) - (x-1)*polyval(poly, x-1))
nprint(max([error(0.75*(1.+n/1000.)) for n in range(1000)]), 12)
```
